```python
import os
import re

# Initialize the list to store results
ga_result = []

# Path to the folder containing the .txt files
folder_path = r'D:\\HOCDIBANTRE\\TULKH\\result\\result_ga'

# Function to extract numbers from filenames
def extract_number(filename):
    match = re.search(r'(\d+)', filename)
    return int(match.group(0)) if match else float('inf')

# Get a sorted list of filenames in the directory based on extracted numbers
sorted_filenames = sorted(os.listdir(folder_path), key=extract_number)

# Iterate over each file in the sorted list
for filename in sorted_filenames:
    file_path = os.path.join(folder_path, filename)
    # Check if the file is a .txt file
    if os.path.isfile(file_path) and filename.endswith('.txt'):
        with open(file_path, 'r') as file:
            # Read the file line by line
            for line in file:
                # Check if the line contains "Objective value"
                if "Objective" in line:
                    # Extract the value after "Objective value: "
                    value = line.split(":")[1].strip()
                    # Append the extracted value to the greedy_result list
                    ga_result.append(value)

# Convert the list of strings to a list of integers
ga_result = [int(x) for x in ga_result]

# Print the greedy_result list to verify
print(ga_result)
```

```python
import matplotlib.pyplot as plt
import numpy as np

list1 = greedy_result
list2 = greedy_local_result
list6 = ga_result
list3 = local_ga_result
list4 = tabu_result
list5 = greedy_local_hill_result

# Split data into four parts
list1_parts = [list1[i:i + 6] for i in range(0, len(list1), 6)]
list2_parts = [list2[i:i + 6] for i in range(0, len(list2), 6)]
list3_parts = [list3[i:i + 6] for i in range(0, len(list3), 6)]
list4_parts = [list4[i:i + 6] for i in range(0, len(list4), 6)]
list5_parts = [list5[i:i + 6] for i in range(0, len(list5), 6)]
list6_parts = [list6[i:i + 6] for i in range(0, len(list6), 6)]

# X-axis labels
labels_parts = [
    ['5_2', '5_3', '10_2', '10_3', '10_5', '20_3'],
    ['20_5', '20_7', '20_10', '50_5', '50_7', '50_10'],
    ['50_20', '50_25', '100_10', '100_50', '200_20', '200_50'],
    ['500_20', '500_50', '700_50', '700_70', '1000_50', '1000_100']
]

# The width of the bars
bar_width = 0.13

# Create the figure and the axes
fig, axs = plt.subplots(2, 2, figsize=(16, 10))

# Flatten the axes array for easy iteration
axs = axs.flatten()

# Plotting each part in its subplot
for i in range(4):
    indices = np.arange(len(list1_parts[i]))
    
    bar1 = axs[i].bar(indices - bar_width * 2.5, list1_parts[i], bar_width, label='greedy')
    bar2 = axs[i].bar(indices - bar_width*1.5, list2_parts[i], bar_width, label='greedy+local')
    bar6 = axs[i].bar(indices - bar_width*0.5, list6_parts[i], bar_width, label='ga')
    bar3 = axs[i].bar(indices + bar_width*0.5, list3_parts[i], bar_width, label='ga+local')
    bar4 = axs[i].bar(indices + bar_width*1.5, list4_parts[i], bar_width, label='tabu')
    bar5 = axs[i].bar(indices + bar_width * 2.5, list5_parts[i], bar_width, label='hill_climbing')
    # Adding labels
    axs[i].set_xlabel('Dataset')
    axs[i].set_ylabel('Objective values')
    axs[i].set_title(f'Comparison of Multiple Heuristics (Part {i+1})')
    axs[i].set_xticks(indices)
    axs[i].set_xticklabels(labels_parts[i])
    axs[i].legend()
    
    # Add labels for each group of bars
    for bar in [bar1, bar2, bar3, bar4, bar5]:
        for rect in bar:
            height = rect.get_height()
            axs[i].annotate('{}'.format(height),
                            xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, 3),  # 3 points vertical offset
                            textcoords="offset points",
                            ha='center', va='bottom', fontsize = 4)

# Adjust layout
plt.tight_layout()

# Show the combined plot
plt.show()
```
